````python
import re
````

````python
def extract_E(filename):
    #psi4out='../ISAPT_IO_files_benchmark26_Psi4_1.1/'+filename+'.out'
    #psi4out='../ISAPT_IO_files_benchmark26_Psi4_1.4/'+filename+'.out'
    psi4out='../ISAPT_IO_files_benchmark26_Psi4_1.4_junccpvdz/'+filename+'.out'
    file = open(psi4out, "r")
    iline=0
    for line in file:
        if re.search(pattern, line):
            iline=iline+1
            eline=line
    if (iline < 1):
        print('ERROR: Can\'t find <',pattern,'> in ',psi4out)
    file.close()
    E=float( eline.split()[position] )
    return E
````

````python
# Globals
pattern='Total SAPT0   '
position=4


mols=[
    'C1L1A1',
    'C1L1A2',
    'C1L1A3',
#   'C2L1A1', 
    'C2L1A2',
    'C2L1A3',
    'C3L1A1',
    'C3L1A2',
    'C3L1A3',
    'C1L2A1',
    'C1L2A2',
    'C1L2A3',
    'C2L2A1',
    'C2L2A2',
    'C2L2A3',
    'C3L2A1',
    'C3L2A2',
    'C3L2A3',
    'C1L3A1',
    'C1L3A2',
    'C1L3A3',
    'C2L3A1',
    'C2L3A2',
    'C2L3A3',
    'C3L3A1',
    'C3L3A2',
    'C3L3A3'
]

Nmol=len(mols)

#file = open('../csv_files/benchmark26_ISAPT_Psi4_1.1.csv', "w")
#file = open('../csv_files/benchmark26_ISAPT_Psi4_1.4.csv', "w")
file = open('../csv_files/benchmark26_ISAPT_Psi4_1.4_junccpvdz.csv', "w")
file.write('ISAPT'+'\n')
for imol in range(Nmol):
    E=extract_E(mols[imol])
    print(imol,E)
    file.write(str(E)+'\n')
    
file.close()    
````
